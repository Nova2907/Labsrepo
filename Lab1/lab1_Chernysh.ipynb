<ol>
    <li>Що таке pip і conda та чим вони відрізняються між собою?</li><br>
    <p>pip — це основний інструмент для встановлення Python-пакетів, який завантажує їх з репозиторію PyPI.</p>
    <p>conda — це більш комплексна система керування пакетами та середовищами. Вона дозволяє встановлювати не лише бібліотеки Python, а й бінарні залежності інших мов (C++, R).
    Ключова відмінність: pip фокусується лише на встановленні бібліотек, тоді як conda керує повним оточенням та сумісністю системних файлів.</p>

<li>Як буде виглядати виконання пунктів 1-3 індивідуального завдання за допомогою Virtualenv?</li><br>
    1. Створення ізольованого середовища: <code>python -m venv myenv</code></p>
    2. Запуск скрипта активації: <code>myenv\Scripts\activate</code></p>
    3. Інсталяція необхідних бібліотек: <code>pip install numpy matplotlib opencv-python jupyter</code></p>
    4. Запуск сервера Jupyter: <code>jupyter notebook</code></p>
